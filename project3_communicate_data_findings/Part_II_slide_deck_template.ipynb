## Investigation Overview

This presentation provide insights on the influence of age on the duration, the influence of gender on duration, the predominance of men and the predominance of subscibers, followed by the most dominant start periode of the day and the start day.

## Dataset Overview

The Ford GoBike System Data is a set of informations about individual rides made
in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('./clean.csv')
df.info()
```

## Distribution of member gender

We can see that member gender is higly dominated by Male.


```python
sb.set_theme(style='whitegrid', font_scale=1.1)
plt.figure(figsize=[14.70, 8.27])
plt.pie(x=df.member_gender.value_counts(),
        labels=df.member_gender.value_counts().index, autopct='%1.1f%%',)
plt.title('Distribution of member gender');
```

## Distributiion of user type

User type is hightly dominated by subscribers

```python
plt.figure(figsize=[14.70, 8.27])
plt.pie(x=df.user_type.value_counts(),
        labels=df.user_type.value_counts().index, autopct='%1.1f%%',)
plt.title('Distribution of user type');
```

## Distribution of start periode of the day
Bike trips mostly start in afternoon



```python
plt.figure(figsize=[14.70, 8.27])
plt.pie(x=df.start_periode_of_day.value_counts(),
        labels=df.start_periode_of_day.value_counts().index, autopct='%1.1f%%',);
plt.title('Start periode of the day');
```

## Distribution of days in the week

Most of the bike trip start on tuesday, wednesday and thursday.



```python
plt.figure(figsize=[14.70, 8.27])
plt.pie(x=df.start_day.value_counts(),
        labels=df.start_day.value_counts().index, autopct='%1.1f%%',)
plt.title('Distribution of days in the week');
```

## Influence of age on duration
Teenagers and adults ride longuer than elders



```python
plt.figure(figsize=[14.70, 8.27])
ax = sb.relplot(data=df, x='member_age', y='duration_sec', aspect=1.7,hue='member_age_category')
plt.xlim(0, 80);
plt.title('Relationship between age and duration');
```

## Proportions of bike share for all trip per member age category

The huge majority of members are against bike share for all the trip.


```python
plt.figure(figsize=[14.70, 8.27])
ax = sb.countplot(data=df, x='bike_share_for_all_trip',
                  hue='member_age_category',)
plt.title('Proportions of bike share for all trip per member age category');
```
